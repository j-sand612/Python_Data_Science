# Visualizing NBA MVP Statistics

Notebook for cleaning and visualizing NBA MVP statistics over the years. Data is from basketball-

```python
import numpy as np 
import pandas as pd 
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
csvLoc = "C:\\Users\\sando\\Documents\\Python_Data_Science\\Datasets\\nba_mvps_stats.csv"
mvpsDf = pd.read_csv(csvLoc, skiprows=range(1))
mvpsDf.head(10)
```

```python
mvpsDf.tail()
```

# Add Year Column for easier indexing and plotting

```python
mvpsDf['Year'] = list(range(2019, 1955, -1))
mvpsDf.set_index('Year', inplace=True)
mvpsDf.head()
```

```python
mvpsDf.index = mvpsDf.index.map(int)
mvpsDf.plot(kind='line')

plt.title('MVP Statistics By Year')
plt.xlabel('Year')
```
